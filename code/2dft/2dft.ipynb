```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fourier():
    img = cv2.imread('./source/stt_screenshot.png')
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    height, width = gray.shape

    '''
    Opencv의 dft함수는 실수로 변환된 영상을 입력받는다.
    따라서 DFT_COMPLEX_OUTPUT을 이용하여 실수로 변환한 영상을 입력한다.
    '''
    dft = cv2.dft(np.float32(gray), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)  # dft로 변환된 주파수를 재배열. 주파수가 0인 부분을 중앙에 위치, 주파수가 커질수록 가장자리에 위치시킨다.
    out = 20*np.log(cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 1]))

    '''
    inverse fourier conversion
    fftshift된 dft_shift를 다시 fftshift에 넣어 변환
    변환된 주파수를 다시 dft함수에 넣어주는데, 이번에는 주파수를 영상으로 만들어 입력
    '''
    inverse_shift = np.fft.fftshift(dft_shift)
    inverse_dft = cv2.dft(inverse_shift, flags=cv2.DFT_INVERSE)
    out2 = cv2.magnitude(inverse_dft[:, :, 0], inverse_dft[:, :, 1])

    plt.subplot(131)
    plt.imshow(gray, cmap='gray')
    plt.title('original')
    plt.subplot(132)
    plt.imshow(out, cmap='gray')
    plt.title('dft')
    plt.subplot(133)
    plt.imshow(out2, cmap='gray')
    plt.title('inverse')

    plt.show()
```
